```python
import pandas as pd
import seaborn as sns
import datetime
```

```python
df = pd.read_excel('FinancialAnalyser_ExpenseReport_2023_09.xlsx')
```

```python
def clean_data(df):
    # Replace gaps forward from the previous valid value in: 'Category'
    df = df.fillna({'Category': df['Category'].ffill()})
    # Filter rows based on column: 'Category'
    df = df[df['Category'] != "Total spend"]
    return df

def extract_year(df):
    # Convert column: 'Date' to datetime
    df['Date'] = pd.to_datetime(df['Date'])
    # Extract year from column: 'Date'
    df['Year'] = df['Date'].dt.year
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```

```python
df_melted = pd.melt(df_clean, id_vars=['Category','Sub-Category'], var_name='Date', value_name='Amount')
```

```python
df_melted = extract_year(df_melted)
df_melted.head()
```
